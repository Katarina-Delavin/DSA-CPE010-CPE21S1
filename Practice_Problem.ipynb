```python
sum_of_even_scores = 0

test_scores = [26, 49, 98, 87, 62, 75]

for score in test_scores:
    if score % 2 == 0:
        sum_of_even_scores += score

print("The sum of the even test scores is:", sum_of_even_scores)
```
